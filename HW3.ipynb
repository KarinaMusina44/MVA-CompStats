## Question 1



Nos paramètres: $$\theta = (\underbrace{ \bar{t_{0}}, \bar{v_{0}} }_{ \theta_{pop} },\underbrace{ \sigma_{\xi},\sigma_{\tau},\sigma }_{ \theta_{i} })$$
Nos variables latentes: $$z=(z_{i},z_{pop})=(\alpha _{i},\tau_{i},t_{0},v_{0})$$
Ecrivons la vraisemblance: 
$$q(y,z,\theta) = q(y \vert z,\theta)\times q((z_{i})_{i}\vert\theta) \times q(z_{pop}\vert\theta)\times q(\theta)$$

Par indépendance des observations: 
$$q((y)_{i,j},(z)_{i},\theta) = \prod_{i=1}^N \left( \left(\prod _{j=1}^{k_{i}} q((y)_{i,j} \vert z,\theta)\right)\times q(z_{i}\vert\theta)\right) \times q(z_{pop}\vert\theta_{pop})\times q(\theta)$$
En prenant le log:
$$\log q((y)_{i,j},(z)_{i},\theta) = \sum_{i=1}^N \left( \sum _{j=1}^{k_{i}}\underbrace{  \log q((y)_{i,j} \vert z,\theta) }_{ A }\right)+\sum_{i=1}^N  \underbrace{ \log q(z_{i}\vert\theta_{i})}_{ B } + \underbrace{ \log q(z_{pop}\vert\theta_{pop}) }_{ C }+\underbrace{ \log q(\theta_{pop}) }_{ D_{1} }+ \underbrace{ \log q(\theta_{i}) }_{ D_{2} }$$
### Pour $A$:
On a  $$y_{ij}\sim \mathcal{N}(d_{i}(t_{ij}),\sigma^2)$$
Donc $$A = -\frac{1}{2}\left( \frac{y_{ij}-d_{i}(t_{ij})}{\sigma} \right)^2 -\log(\sigma) + const$$
### Pour B: 
$$z_{i}= \begin{pmatrix}
\alpha_{i} \\
\tau_{i}
\end{pmatrix} = \begin{pmatrix}
\exp(\xi_{i}) \text{ avec } \xi_{i}\sim \mathcal{N}(0,\sigma_{\xi}^2) \\
\tau_i\sim \mathcal{N}(0,\sigma_{\tau}^2)
\end{pmatrix}$$
On obtient la somme des deux termes en passant au log:
$$B = -\frac{1}{2}\left( \frac{\xi_{i}^2}{\sigma_{\xi}^2} +\frac{\tau_{i}^2}{\sigma_{\tau}^2} \right) -\log(\sigma_{\tau})-\log(\sigma_{\xi}) + const$$
### Pour C: 
$$t_{0}\sim \mathcal{N}(\bar{t_{0}},\sigma_{t_{0}}^2)$$
$$v_{0}\sim \mathcal{N}(\bar{v_{0}},\sigma_{v_{0}}^2)$$
Ainsi 
$$C= -\frac{1}{2}\left( \frac{(t_{0}-\bar{t_{0}})^2}{\sigma_{t_{0}}^2} +\frac{(v_{0}-\bar{v_{0}})^2}{\sigma_{v_{0}}^2} \right) + const$$

### Pour $D_{1}$:

$$\bar{t}_{0}\sim \mathcal{N}(\bar{\bar{t_{0}}},s_{t_{0}}^2)$$
$$\bar{v}_{0}\sim \mathcal{N}(\bar{\bar{v_{0}}},s_{v_{0}}^2)$$
$$D_{1}= -\frac{1}{2}\left( \frac{\bar{(t_{0}}-\bar{\bar{t_{0}}})^2}{s_{t_{0}}^2} +\frac{\bar{(v_{0}}-\bar{\bar{v_{0}} })^2}{s_{v_{0}}^2} \right) + const$$

### Pour $D_{2}$:

$$
\sigma_\xi^2 \sim W^{-1}(v_\xi,m_\xi) ;\quad
\sigma_\tau^2 \sim W^{-1}(v_\tau,m_\tau) ;\quad
\sigma^2 \sim W^{-1}(v,m)
$$

$$
f_{W^{-1}}(\sigma^2)
= \frac{1}{\Gamma\left(\frac{m}{2}\right)} \frac{1}{\sigma^2}
\left(\frac{v}{\sigma \sqrt{ 2 }}\right)^m
\exp\left(-\frac{v^2}{2\sigma^2}\right).
$$
Ainsi:
$$\begin{aligned} D_{2}=-\log(\sigma^2)-\frac{v^2}{2\sigma^2}+m\log\left( \frac{v}{\sigma} \right) \\
-\log(\sigma_{\xi}^2)-\frac{v_{\xi}^2}{2\sigma_{\xi}^2}+m_{\xi}\log\left( \frac{v_{\xi}}{\sigma_{\xi}} \right)
\\
-\log(\sigma_{\tau}^2)-\frac{v_{\tau}^2}{2\sigma_{\tau}^2}+m_{\tau}\log\left( \frac{v_{\tau}}{\sigma_{\tau}} \right)\end{aligned}$$



### On obtient donc:
$$\begin{aligned}
\log q((y)_{i,j},(z)_{i},\theta) =
 \sum_{i=1}^N \left( \sum _{j=1}^{k_{i}}\underbrace{  -\frac{1}{2}\left( \frac{y_{ij}-d_{i}(t_{ij})}{\sigma} \right)^2 -\log(\sigma) }_{ A }\right)

\\
+\sum_{i=1}^N  \underbrace{ -\frac{1}{2}\left( \frac{\xi_{i}^2}{\sigma_{\xi}^2} +\frac{\tau_{i}^2}{\sigma_{\tau}^2} \right) -\log(\sigma_{\tau})-\log(\sigma_{\xi}) }_{ B }

\\ -\log(\sigma^2)-\frac{v^2}{2\sigma^2}+m\log\left( \frac{v}{\sigma} \right) \\
-\log(\sigma_{\xi}^2)-\frac{v_{\xi}^2}{2\sigma_{\xi}^2}+m_{\xi}\log\left( \frac{v_{\xi}}{\sigma_{\xi}} \right)
\\
-\underbrace{\log(\sigma_{\tau}^2)-\frac{v_{\tau}^2}{2\sigma_{\tau}^2}+m_{\tau}\log\left( \frac{v_{\tau}}{\sigma_{\tau}} \right) }_{ D_{2} } 

\\ \underbrace{ -\frac{1}{2}\left( \frac{(t_{0}-\bar{t_{0}})^2}{\sigma_{t_{0}}^2} +\frac{(v_{0}-\bar{v_{0}})^2}{\sigma_{v_{0}}^2} \right) }_{ C }
\\\underbrace{ -\frac{1}{2}\left( \frac{\bar{(t_{0}}-\bar{\bar{t_{0}}})^2}{s_{t_{0}}^2} +\frac{\bar{(v_{0}}-\bar{\bar{v_{0}} })^2}{s_{v_{0}}^2} \right) }_{ D_{1} }
\\
+ const
 \end{aligned}$$
 En notant $\frac{1}{N}\sum_{i=1}^Nk_{i}=K$ 
 $$\begin{aligned}
\log q((y)_{i,j},(z)_{i},\theta) =
 -\frac{1}{2\sigma^2}\sum_{i=1}^N  \sum _{j=1}^{k_{i}} \left( y_{ij}-d_{i}(t_{ij})\right)^2 -K\log(\sigma)

\\
+\sum_{i=1}^N  -\frac{1}{2}\left( \frac{\xi_{i}^2}{\sigma_{\xi}^2} +\frac{\tau_{i}^2}{\sigma_{\tau}^2} \right) -N\log(\sigma_{\tau}\sigma_{\xi})

\\
-\log(\sigma^2)-\frac{v^2}{2\sigma^2}+m\log\left( \frac{v}{\sigma} \right)
\\
-\log(\sigma_{\xi}^2)-\frac{v_{\xi}^2}{2\sigma_{\xi}^2}+m_{\xi}\log\left( \frac{v_{\xi}}{\sigma_{\xi}} \right)
\\
-\log(\sigma_{\tau}^2)-\frac{v_{\tau}^2}{2\sigma_{\tau}^2}+m_{\tau}\log\left( \frac{v_{\tau}}{\sigma_{\tau}} \right) 

\\  -\frac{1}{2}\left( \frac{(t_{0}-\bar{t_{0}})^2}{\sigma_{t_{0}}^2} +\frac{(v_{0}-\bar{v_{0}})^2}{\sigma_{v_{0}}^2} \right) 
\\ -\frac{1}{2}\left( \frac{\bar{(t_{0}}-\bar{\bar{t_{0}}})^2}{s_{t_{0}}^2} +\frac{\bar{(v_{0}}-\bar{\bar{v_{0}} })^2}{s_{v_{0}}^2} \right)
\\
+ const
 \end{aligned}$$
 
La famille expotentielle s'écrit sous la forme:
$$\log q(y,z,\theta) = -\phi(\theta) + < S(y,z),\psi(\theta)> $$

On va regrouper dans $S$ les termes avec $z$ et $y$:
$$S(y,z)=\begin{pmatrix}\frac{1}{KN}
\sum_{i=1}^N \sum _{j=1}^{k_{i}} \left( y_{ij}-d_{i}(t_{ij})\right)^2 \\
\frac{1}{N}\sum_{i=1}^N\xi_{i}^2 = \sum_{i=1}^N z_{i1}\\
\frac{1}{N}\sum_{i=1}^N\tau_{i}^2 = \sum_{i=1}^N z_{i2} \\
t_{0} = z_{pop1} \\
v_{0} = z_{pop2}
\end{pmatrix}$$
On trouve ainsi $\psi(\theta)$ en complétant le produit scalaire
$$\psi(\theta)=\begin{pmatrix}
-\frac{KN}{2\sigma^2} \\
-\frac{N}{2\sigma_{\xi}^2} \\
-\frac{N}{2\sigma_{\tau}^2} \\
\frac{\bar{t_{0}}}{\sigma_{t_{0}}^2} \\
\frac{\bar{v_{0}}}{\sigma_{v_{0}}^2}
\end{pmatrix}$$
Le reste des termes vont dans $\phi$:
$$\begin{aligned}
-\phi(\theta)= -K\log(\sigma) -N\log(\sigma_{\tau}\sigma_{\xi})\\
-\log(\sigma^2)-\frac{v^2}{2\sigma^2}+m\log\left( \frac{v}{\sigma} \right)
\\
-\log(\sigma_{\xi}^2)-\frac{v_{\xi}^2}{2\sigma_{\xi}^2}+m_{\xi}\log\left( \frac{v_{\xi}}{\sigma_{\xi}} \right)
\\
-\log(\sigma_{\tau}^2)-\frac{v_{\tau}^2}{2\sigma_{\tau}^2}+m_{\tau}\log\left( \frac{v_{\tau}}{\sigma_{\tau}} \right)
\\
 - \frac{\bar t_0^2}{2\sigma_{t_0}^2}- \frac{\bar v_0^2}{2\sigma_{v_0}^2} 
\\
-\frac{1}{2}\left( \frac{\bar{(t_{0}}-\bar{\bar{t_{0}}})^2}{s_{t_{0}}^2} +\frac{\bar{(v_{0}}-\bar{\bar{v_{0}} })^2}{s_{v_{0}}^2} \right)
\\
+ const
\end{aligned}$$


et $c(y,z) = -\frac{t_{0}^2}{2\sigma^2_{t_{0}}}-\frac{v_{0}^2}{2\sigma^2_{v_{0}}}$

```python
#z_i
tau = [0] * N
alpha = [0] * N
xi = [0] * N
for i in range(N):
    tau[i] = np.random.normal(0,sigma_tau)
    xi[i] = np.random.normal(0,sigma_xi)
    alpha[i] = np.exp(xi[i])
```

```python
z = np.zeros(2 + 2*N)
z[0] = t0
z[1] = v0
for i in range(N):
    z[2+2*i]     = xi[i]
    z[2+2*i + 1] = tau[i]
```

```python
t = {}
for i in range(N):
    t_i = []
    current_t = 0
    for j in range(ki[i]):
        current_t += np.random.uniform()
        t_i.append(current_t)
    t[i] = t_i
```

```python
#traj
def d(t,p0,v0,t0):
    return (p0 + v0*(t-t0))


def di(i, t, z):
    t0 = z[0]
    v0 = z[1]
    alpha_i = np.exp(z[2 + 2*i])
    tau_i = z[2 + 2*i + 1]
    t_aux = alpha_i * (t - t0 - tau_i) + t0
    return d(t_aux, p0, v0, t0)
```

# Question 3

On va utiliser l'algorithme de Metropolis-Hastings pour échantillonner $z = (t_{0},v_{0},\xi_{i},\tau_{i})$ a posteriori. 
Pour cela on utilisera la distribution de proposition qui est une gaussienne multivariée  $\mathcal{N}(z,\sigma_{prop})$
On va donc avoir besoin de fixer $\sigma_{prop}$ et de pouvoir calculer $\alpha = min\left( 1,\frac{\pi(z^\star)}{\pi(z^{(k)})} \right)$ avec $\pi(z) = q(z\vert y,\theta) = \frac{q(z,y,\theta)}{q(y,\theta)}$

Mais en fait ce qui nous intéresse c'est le ratio:

On a:
$$\frac{\pi(z^\star)}{\pi(z^{(k)})}=\frac{q(z^\star,y,\theta)}{q(z^{(k)},y,\theta)}$$
donc $$\log(\frac{\pi(z^\star)}{\pi(z^{(k)})})= c(y,z^\star)-c(y, z^{(k)})+<S(y,z^\star),\psi(\theta)>- <S(y,z^{(k)}),\psi(\theta)>$$
puis $$\log(\frac{\pi(z^\star)}{\pi(z^{(k)})})= c(y,z^\star)-c(y, z^{(k)})+<S(y,z^\star)- S(y,z^{(k)}),\psi(\theta)>$$


```python
Ki_sum = np.mean(ki)
```

```python
def c(y,z):
    t0 = z[0]
    v0 = z[1]
    return - (t0**2/(2*sigma_t0**2) + v0**2/(2*sigma_v0**2))
```

```python
def S(y, z):
    
    t0 = z[0]
    v0 = z[1]

    sse = 0
    for i in range(N):
        for j in range(ki[i]):
            sse += (y[i][j] - di(i, t[i][j], z))**2
    S1 = sse / (Ki_sum * N)
    xi = z[2 + 2*np.arange(N)]
    tau = z[2 + 2*np.arange(N) + 1]
    S2 = np.mean(xi**2)
    S3 = np.mean(tau**2)
    S4 = t0
    S5 = v0

    return np.array([S1, S2, S3, S4, S5])
```

```python
def psi(theta): 
    t0bar, v0bar, sigma_xi, sigma_tau, sigma = theta
    return np.array([-(Ki_sum*N)/(2*sigma**2), -N/(2*sigma_xi**2), -N/(2*sigma_tau**2), t0bar/(sigma_t0**2), v0bar/(sigma_v0**2)])
```

```python
def scalar_prod_S(y,z,theta):
    return S(y,z)@np.transpose(psi(theta))
```

```python

def SRWMH(z0,sigma_prop,theta,maxiter=100):
    z = [z0]
    z_k = z0
    acc_list =[]
    
    for k in range(maxiter):
        #proposal
        z_prop = np.random.multivariate_normal(z_k,sigma_prop)
        #acception/rejection
        log_ratio = c(y,z_prop)-c(y,z_k)+scalar_prod_S(y,z_prop,theta)-scalar_prod_S(y,z_k,theta)
        acc = min(1,np.exp(log_ratio))
        acc_list.append(acc)
        # update (ou non)
        u = np.random.uniform()
        if u<acc:
            z_k = z_prop
        z.append(z_k)
    return(z,acc_list)
```

```python
# On fait la marche symmetrique
step_t0  = 0.0012
step_v0  = 0.0012
step_xi  = 0.0012
step_tau = 0.0012

steps = np.array([step_t0, step_v0] + [step_xi, step_tau] * N)
sigma_prop = np.diag(steps**2)
```

```python
maxiter = 400

z_chain, acc_list = SRWMH(z, sigma_prop, true_theta, maxiter)

print("Taux d'acceptation moyen:", np.mean(acc_list))
```

$\theta = (\underbrace{ \bar{t_{0}}, \bar{v_{0}} }_{ \theta_{pop} },\underbrace{ \sigma_{\xi},\sigma_{\tau},\sigma }_{ \theta_{i} })$
Cherchons  $\theta^{(k)} = \underset{ \theta \in \Theta }{ argmax}\{ -\phi(\theta)+\langle S_{k},\psi(\theta)\rangle\}$
En dérivant l'expression (gradient): 
$$
\nabla_\theta \big(-\phi(\theta)+\langle S_k,\psi(\theta)\rangle\big)
=
\begin{pmatrix}
- \frac{\bar{t}_{0}}{\sigma_{t_{0}}^{2}}
- \frac{\bar{t}_{0}}{s_{t_{0}}^{2}}
+ \frac{\bar{\bar{t_{0}}}}{s_{t_{0}}^{2}}
+ \frac{S_{k4}}{\sigma_{t_{0}}^{2}}
\\
- \frac{\bar{v}_{0}}{\sigma_{v_{0}}^{2}}
- \frac{\bar{v}_{0}}{s_{v_{0}}^{2}}
+ \frac{\bar{\bar{v_{0}}}}{s_{v_{0}}^{2}}
+ \frac{S_{k5}}{\sigma_{v_{0}}^{2}}
\\
-\frac{N + m_{\xi} + 2}{\sigma_{\xi}}
+ \frac{v_{\xi}^{2} + S_{k2}N}{\sigma_{\xi}^{3}}
\\
-\frac{N + m_{\tau} + 2}{\sigma_{\tau}}
+ \frac{v_{\tau}^{2} + S_{k3}N}{\sigma_{\tau}^{3}}
\\
-\frac{K + m + 2}{\sigma}
+ \frac{v^{2} + S_{k1}KN}{\sigma^{3}}
\end{pmatrix}.
$$


On obtient les mises à jour suivantes :
$$
\begin{aligned}
\bar t_0^{(k)}= \frac{\frac{\bar{\bar {t_0}}}{s_{t_0}^2}
      + \frac{S_{k4}}{\sigma_{t_0}^2}}
     {\frac{1}{s_{t_0}^2}
      + \frac{1}{\sigma_{t_0}^2}}
\\
\bar v_0^{(k)}= \frac{ \frac{\bar{\bar v_0}}{s_{v_0}^2}
      + \frac{S_{k5}}{\sigma_{v_0}^2}}
     { \frac{1}{s_{v_0}^2}
      + \frac{1}{\sigma_{v_0}^2}}
\\
\sigma_\xi^{(k)}= \sqrt{ \frac{v_\xi^2 + S_{k2}N}{N + m_\xi + 2} }
\\
\sigma_\tau^{(k)}= \sqrt{ \frac{v_\tau^2 + S_{k3}N}{N + m_\tau + 2} }
\\
\sigma^{(k)}= \sqrt{ \frac{v^2 + S_{k1}KN}{K + m + 2} }
\end{aligned}
$$

```python
def maximize(S):
    #t0bar
    num = (t0barbar/(s_t0**2)+S[3]/(sigma_t0**2))
    den = (1/(s_t0**2)+1/(sigma_t0**2))
    t0bar = num/den
    #v0bar
    num = (v0barbar/(s_v0**2)+S[4]/(sigma_v0**2))
    den = (1/(s_v0**2)+1/(sigma_v0**2))
    v0bar = num/den
    
    # sigma_xi
    sigma_xi = np.sqrt((v_xi**2+S[1]*N)/(N+2+m_xi))

    # sigma_tau
    sigma_tau = np.sqrt((v_tau**2+S[2]*N)/(N+2+m_tau))

    # sigma
    sigma = np.sqrt((v**2+S[0]*Ki_sum*N)/(Ki_sum*N+2+m))

    
    return np.array([t0bar,v0bar,sigma_xi,sigma_tau,sigma])
```

```python
def MCMC_SAEM(y,theta0,burn_in=300, maxiter = 600, alpha_eps = 0.8):
    theta = theta0 
    theta_list = [theta0]
    z = np.zeros(2*N+2)
    current_S = np.zeros(5)
    for k in range(1,maxiter+1):
        print(f'Iter{k}: {theta}')
        # calcul d'epsilon
        if k<=burn_in:
            epsilon = 1
        else:
            epsilon = (k-burn_in)**(-alpha_eps)
        #sample    
        z_list, acc = SRWMH(z,sigma_prop,theta,maxiter=20)
        z = z_list[-1]
        #update
        current_S +=epsilon*(S(y,z)-current_S)
        theta = maximize(current_S)
        theta_list.append(theta)
    return theta_list
    
```

```python
def MH_Gibbs(z0, sigma_prop, theta, maxiter=100):  # sigma_prop scalaire cette fois ci
    z = [z0]
    z_k = z0
    acc_list = []


    log_pi_zk = c(y, z_k) + scalar_prod_S(y, z_k, theta)

    for k in range(maxiter): 
        for i in range(N):  # sample xi_i and tau_i
            #xi_i 
            xi_i_prop = np.random.normal(z_k[2 + 2*i], sigma_prop)
            z_prop = z_k.copy()
            z_prop[2 + 2*i] = xi_i_prop

            log_pi_prop = c(y, z_prop) + scalar_prod_S(y, z_prop, theta)
            log_ratio = log_pi_prop - log_pi_zk

            acc = min(1, np.exp(log_ratio))
            acc_list.append(acc)
            u = np.random.uniform()
            if u < acc:
                z_k = z_prop
                log_pi_zk = log_pi_prop

            #  tau_i 
            tau_i_prop = np.random.normal(z_k[2 + 2*i + 1], sigma_prop)
            z_prop = z_k.copy()
            z_prop[2 + 2*i + 1] = tau_i_prop

            log_pi_prop = c(y, z_prop) + scalar_prod_S(y, z_prop, theta)
            log_ratio = log_pi_prop - log_pi_zk

            acc = min(1, np.exp(log_ratio))
            acc_list.append(acc)
            u = np.random.uniform()
            if u < acc:
                z_k = z_prop
                log_pi_zk = log_pi_prop
        
        #  sample t0 
        t0_prop = np.random.normal(z_k[0], sigma_prop)
        z_prop = z_k.copy()
        z_prop[0] = t0_prop

        log_pi_prop = c(y, z_prop) + scalar_prod_S(y, z_prop, theta)
        log_ratio = log_pi_prop - log_pi_zk

        acc = min(1, np.exp(log_ratio))
        acc_list.append(acc)
        u = np.random.uniform()
        if u < acc:
            z_k = z_prop
            log_pi_zk = log_pi_prop
            
        #  sample v0 
        v0_prop = np.random.normal(z_k[1], sigma_prop)
        z_prop = z_k.copy()
        z_prop[1] = v0_prop

        log_pi_prop = c(y, z_prop) + scalar_prod_S(y, z_prop, theta)
        log_ratio = log_pi_prop - log_pi_zk

        acc = min(1, np.exp(log_ratio))
        acc_list.append(acc)
        u = np.random.uniform()
        if u < acc:
            z_k = z_prop
            log_pi_zk = log_pi_prop
            
    return (z_k, acc_list)
```

```python
def HMwG_SAEM(y,theta0,burn_in=100, maxiter = 200, alpha_eps = 0.8):
    theta = theta0 
    theta_list = [theta0]
    z = np.zeros(2*N+2)
    current_S = np.zeros(5)
    for k in range(1,maxiter+1):
        print(f'Iter{k}: {theta}')
        if k<=burn_in:
            epsilon = 1
        else:
            epsilon = (k-burn_in)**(-alpha_eps)
        z, acc = MH_Gibbs(z,sigma_prop_gibbs,theta,maxiter=5) #la seule chose qui change par rapport au code precedent est le sampler
        current_S +=epsilon*(S(y,z)-current_S)
        theta = maximize(current_S)
        theta_list.append(theta)
    return theta_list
```

```python
def MH_Gibbs_bloc(z0, sigma_prop, theta, maxiter=100):  # sigma_prop scalaire
    z = [z0]
    z_k = z0
    acc_list = []

    log_pi_zk = c(y, z_k) + scalar_prod_S(y, z_k, theta)

    for k in range(maxiter):
        # bloc (xi_i, tau_i)
        for i in range(N):
            current_block = z_k[2 + 2*i : 2 + 2*i + 2]
            cov_block = np.diag([sigma_prop**2,sigma_prop**2])
            prop_block = np.random.multivariate_normal(current_block, cov_block)
            z_prop = z_k.copy()
            z_prop[2 + 2*i : 2 + 2*i + 2] = prop_block

            log_pi_prop = c(y, z_prop) + scalar_prod_S(y, z_prop, theta)
            log_ratio = log_pi_prop - log_pi_zk

            acc = min(1, np.exp(log_ratio))
            acc_list.append(acc)
            u = np.random.uniform()
            if u < acc:
                z_k = z_prop
                log_pi_zk = log_pi_prop

        # bloc (t0, v0)
        current_block = z_k[0:2]
        cov_block = (sigma_prop**2) * np.eye(2)
        prop_block = np.random.multivariate_normal(current_block, cov_block)
        z_prop = z_k.copy()
        z_prop[0:2] = prop_block

        log_pi_prop = c(y, z_prop) + scalar_prod_S(y, z_prop, theta)
        log_ratio = log_pi_prop - log_pi_zk

        acc = min(1, np.exp(log_ratio))
        acc_list.append(acc)
        u = np.random.uniform()
        if u < acc:
            z_k = z_prop
            log_pi_zk = log_pi_prop

    return (z_k, acc_list)
```

Ceci nous donne l'algorithme suivant:
On part de $(X_{0},Y_{0})$
 On répète: 
1. Echantillonnage de $X_{n}$ suivant $f_{X\vert Y}(.,Y_{n-1})$
2. Echantillonnage de $Y_{n}$ suivant $f_{Y\vert X}(X_{n-1},.)$

Pour un nombre d'itérations suffisamment grand, on obtient la distribution désirée.



```python
def gibbs_sampler(x0,y0,n_iter):
    samples = []
    last_x = x0
    last_y= y0
    samples.append([last_x,last_y])
    for i in range(n_iter):
        last_x =  np.random.normal(loc=0, scale=1/np.sqrt(last_y)) 
        last_y = np.random.gamma(shape=5/2, scale=1/(last_x**2/2 + 2)) 
        samples.append([last_x,last_y])
    return samples
```
